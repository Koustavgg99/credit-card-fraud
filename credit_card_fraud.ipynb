```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'creditcard_2023.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Print the contents of the DataFrame
print(df)
```

```python
df = df.dropna()

# Remove duplicate rows
df = df.drop_duplicates()

# Print the cleaned DataFrame
print(df)

# Optionally, save the cleaned DataFrame back to a new CSV file
#cleaned_csv_file_path = 'path/to/your/cleaned_file.csv'
#df.to_csv(cleaned_csv_file_path, index=False)
```

```python
x = credit_card.drop(columns= 'Class')
y= credit_card[['Class']]
```

```python
# run Isolation forest model
from sklearn.ensemble import IsolationForest
is_forest = IsolationForest(n_estimators= 300, max_samples= 100000, contamination = 0.2, bootstrap= True, random_state = 0)
otliers_prediction = is_forest.fit_predict(x)
otliers_prediction
```

```python
anomaly_score = (is_forest.decision_function(x))
anomaly_score
```

```python
import numpy as np
# set a threshold 10%
anomaly_score = -1*anomaly_score
percentile_threshold = np.percentile(anomaly_score, 90)
percentile_threshold
```

```python
anomaly_score
```

```python
# filter the anomalies by the threshold
outlier_threshold = np.where(anomaly_score>= percentile_threshold)[0]
outlier_threshold
len(outlier_threshold)
```

```python
# drop the outliers
banking_version_one = credit_card.copy(deep= True)
banking_version_one.drop(index= outlier_threshold, inplace= True)
```

```python
# Check the outliers
fig, axes = plt.subplots(4,8,figsize = (20,20))
axes = axes.flatten()
for i, column in enumerate(banking_version_one.columns):
    sns.boxplot(banking_version_one[column], ax= axes[i])
    axes[i].set_title(f'Box Plot for {column}')

plt.tight_layout()
plt.show()
```

```python
banking_version_one['Class'].value_counts()
```

```python
x=  banking_version_one.drop(columns= 'Class')
y= banking_version_one[['Class']]
len(x) , len(y)
```

```python
# implement a logisitic regression model
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score   , precision_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
lg_model = LogisticRegression(random_state= 0)
parameters = {'penalty': ['l1', 'l2'], 'C': [30,40,50], 'solver' : ['saga', 'lbfgs']}
lg_grid_model = GridSearchCV(estimator= lg_model, param_grid = parameters, cv = 5, scoring= 'recall',)
lg_grid_model.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
y_pred1=lg_grid_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score   , precision_score,f1_score
print(accuracy_score(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
print(recall_score(y_test,y_pred1))
print(f1_score(y_test,y_pred1))
```

```python
# the outcomes of the logistic regression
lg_grid_model.best_score_, lg_grid_model.best_params_, lg_grid_model.cv_results_
```

```python
# implement a random forest model
import warnings
warnings.filterwarnings("ignore")
from sklearn.ensemble import RandomForestClassifier
rn_model = RandomForestClassifier(random_state=0, bootstrap= True)
parameters = {'criterion': ['entropy', 'gini'], 'max_depth': [6,9,10], 'max_samples' : [18000,20000]}
rn_grid_model = GridSearchCV(estimator= rn_model, param_grid = parameters, cv = 5, scoring= 'recall')
rn_grid_model.fit(x,y)
```

```python
# Random forest model outcomes
rn_grid_model.best_score_, rn_grid_model.best_params_
```

```python
y_pred2=rn_grid_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score   ,f1_score, precision_score
print(accuracy_score(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
print(recall_score(y_test,y_pred2))
print(f1_score(y_test,y_pred2))
```

```python
final_data =pd.DataFrame({'Models':['LR','RN'],'ACC':[accuracy_score(y_test,y_pred1)*100,accuracy_score(y_test,y_pred2)*100],
                                                     'Precision':[precision_score(y_test,y_pred1)*100,precision_score(y_test,y_pred2)*100],
                          'Recall':[recall_score(y_test,y_pred1)*100,recall_score(y_test,y_pred2)*100],
                                'F1':[f1_score(y_test,y_pred1)*100,f1_score(y_test,y_pred2)*100]})
```
